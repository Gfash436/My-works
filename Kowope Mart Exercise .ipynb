```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from scipy.stats import skew

pd.set_option('display.max_columns', None)

import copy
```

```python
#Reading the data files
train = pd.read_csv('train.csv', index_col='Applicant_ID')

train.head()
```

```python
test = pd.read_csv('test.csv', index_col='Applicant_ID')

test.head()
```

```python
train.isna().sum()#Check if missing values exists
```

```python
#Handling Missing Values
train = train.fillna(-999)
test = test.fillna(-999)
```

```python
train.isnull().sum()
```

```python
train.isnull().sum().any()
```

```python
#Join train and test dataset
data = pd.concat([train.drop("default_status", axis=1),test], axis=0) 
```

```python
#Handling the Categorical data with one-hot encoding
data = pd.get_dummies(data=data)
```

```python
data.head()
```

```python
#Seperating the Target from the Predictors
y = train[["default_status"]]

train.head(2)
```

```python
data.shape
```

```python
X = data.iloc[:56000,:]
X_test = data.iloc[-24000:, :]
```

```python
X.shape, X_test.shape, y.shape
```

```python
# Scaling the Data

from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
```

```python
X = scaler.fit_transform(X)
X_test = scaler.transform(X_test)
```

```python
#Building the Model and Testing

from sklearn.model_selection import cross_val_score
from lightgbm import LGBMClassifier

model = LGBMClassifier()
cv_score = cross_val_score(model, X, y, scoring="roc_auc")
```

```python
print(cv_score)

cv_score.mean()
```

```python
#Fitting the model
model.fit(X, y)
```

```python
test_pred = model.predict_proba(X_test)[:, 1]
```
