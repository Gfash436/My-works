```python
#Importing Other Necessary Libraries
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import cross_val_score
```

```python
%pip install optuna
```

```python
#Parameter Optimization
import optuna

def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 500,750)
    max_depth = round(trial.suggest_float('max_depth', 1, 32, log=True))
    border_count = trial.suggest_int('border_count', 150, 200)
    subsample= trial.suggest_uniform('subsample', 0.1, 1.0)
    learning_rate = trial.suggest_loguniform('learning_rate', 0.01, 1.0)
    random_strength = trial.suggest_float('random_strength',  0.05,0.1)
    scale_pos_weight = trial.suggest_float('scale_pos_weight', 0.1, 1.0)
    bagging_temperature = trial.suggest_float('bagging_temperature', 0.5,1.0)

    clf = CatBoostClassifier(
        n_estimators=n_estimators, max_depth=max_depth, 
        bagging_temperature=bagging_temperature, scale_pos_weight=scale_pos_weight,
        border_count=border_count, subsample=subsample, learning_rate=learning_rate,
        random_strength=random_strength)
    
    return cross_val_score(
        clf,X,y, n_jobs=-1,scoring='roc_auc',cv=StratifiedKFold(n_splits=3,shuffle=True,random_state=360)).mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)

trial = study.best_trial

print('Accuracy: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
best_cat_params = trial.params
```

```python
best_cat_params["max_depth"] = int(best_cat_params["max_depth"])
```

```python
# Creating instances for both Kfold and StratifiedKfold
splits = 25

skfold = StratifiedKFold(n_splits=splits , random_state=419, shuffle=True)
```

```python
#Models used for Voting
cat = CatBoostClassifier(**best_cat_params,random_state=419, verbose=False, thread_count=-1)
rfc = RandomForestClassifier(n_estimators=400, random_state=419, verbose=False, n_jobs=-1)
lgb = LGBMClassifier(n_jobs=-1, n_estimators=400)
```

```python
from sklearn.metrics import roc_auc_score
def stratified_vote():
    scores, preds = [], []
    i = 1
    for train_split, val_split in skfold.split(X, y):
        x_train, x_test, y_train, y_test = X.iloc[train_split],X.iloc[val_split], y.iloc[train_split],y.iloc[val_split]
        
        vote = VotingClassifier(estimators=[('model1',cat),('model2',rfc),('model3',lgb)], voting='soft')
        vote.fit(x_train, y_train)
        
        #To see how the model performs after splitting
        score = roc_auc_score(y_test,vote.predict_proba(x_test)[:,1])
        pred = vote.predict_proba(X_test)[:,1]
        scores.append(score)
        preds.append(pred)
        print('Roc for {} split: '.format(i), score)
        i += 1
    print('final absolute Roc: ', np.mean(scores))
    final_predictions = np.mean(preds, axis=0)
    return final_predictions

final_predictions = stratified_vote()
        
```
